```python
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Define model and tokenizer with a language model
model_name = "roberta-base"
tokenizer_name = "roberta-base"

# Import model and tokenizer
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

# Train data and label them
train_data = ["This is a sentence about dogs.", "Cats are also very cute animals.", "I love to read books."]
labels = [1, 1, 0]
label = torch.tensor(labels)
# Preparation data
encoded_train_data = tokenizer(train_data, padding=True, truncation=True, max_length=512, return_tensors="pt")

# Define a function for train data
def train(model, encoded_train_data, labels):
    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
    criterion = torch.nn.CrossEntropyLoss()

    for epoch in range(10):
        outputs = model(**encoded_train_data)
        loss = criterion(outputs.logits, label)

        # Update parameters
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        print(f"Epoch {epoch + 1}: Loss {loss.item():.4f}")

# Final train
train(model, encoded_train_data, labels)

# Save final model
torch.save(model.state_dict(), "roberta.pt")
```
